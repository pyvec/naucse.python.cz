```python
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans

%matplotlib inline
```

# K-means

```python
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=0.70, random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1]);
plt.savefig("kmeans_empty.png")
```

```python
kmeans = KMeans(n_clusters=3, max_iter=1, init="random", n_init=1, precompute_distances=False, random_state=0)
kmeans.fit(X)
y_1 = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_1, cmap=cm.viridis);
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c="blue", s=100);
plt.savefig("kmeans_1.png")
```

```python
kmeans = KMeans(n_clusters=3, max_iter=2, init="random", n_init=1, precompute_distances=False, random_state=0)
kmeans.fit(X)
y_2 = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_2, cmap=cm.viridis);
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c="blue", s=100);
plt.savefig("kmeans_2.png")
```
