## Fibonacciho posloupnost

```python
pocet = 20
a = 0
b = 1

print(a, b, sep=", ", end=", ")

for i in range(pocet):
    dalsi_v_rade = a + b
    a = b
    b = dalsi_v_rade

    print(dalsi_v_rade, end=", ")
```
